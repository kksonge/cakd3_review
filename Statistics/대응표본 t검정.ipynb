### 2집단 데이터에 대한 t검정 - 대응표본 t검정
* 2개의 변수 사이에서 평균값에 차이가 있는지 판단
* 약을 먹기 전과 후의 체온 차이와 같이, 같은 대상을 다른 조건으로 2번 측정해서 그 차이를 보는 경우 사용
* '차잇값이 0과 유의미한 차이가 있는지'를 가지고 1집단 t검정 실행

```python
# 수치 계산을 위한 라이브러리
import numpy as np
import pandas as pd
from scipy import stats

# 그래프를 그리기 위한 라이브러리
from matplotlib import pyplot as plt
import seaborn as sns
sns.set()

# 표시 자릿수 지정
%precision 3

# 그래프를 주피터 노트북에 그리기 위한 설정
%matplotlib inline
```

```python
paired_test_data = pd.read_csv('/content/drive/MyDrive/python study/Statistics/dataset/3-9-1-paired-t-test.csv')
print(paired_test_data)
```

### t검정 구현
귀무가설 : 약을 먹기 전과 후의 체온이 변하지 않는다.  
대립가설 : 약을 먹기 전과 후의 체온이 다르다.  

p값이 0.05(유의수준 5%)보다 작으면 귀무가설을 기각하여 약을 먹으면 체온이 유의미한 수준으로 변화한다고 볼 수 있다.

```python
# 약을 먹기 전과 후의 표본평균
before = paired_test_data.query('medicine== "before"')['body_temperature']
after = paired_test_data.query('medicine == "after"')['body_temperature']

# 배열형으로 변환
before = np.array(before)
after = np.array(after)

# 차이 계산
diff = after - before
diff
```

```python
stats.ttest_1samp(diff, 0)
```

```python
# stats.ttest_rel 함수 사용
stats.ttest_rel(after, before)

## p값이 0.05보다 작으므로 약을 먹기 전과 후의 체온은 유의미한 차이가 있다고 할 수 있다.
```

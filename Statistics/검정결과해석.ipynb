#### 1. p 값이 0.05 이하일 때의 결과 작성  
p 값이 0.05 이하일 때는 '유의미한 차이가 있다'고 기재한다.

#### 2. p 값이 0.05보다 클 때의 결과 작성    
p 값이 0.05보다 클 때는 '유의미한 차이가 있다고 말할 수 없다'고 기재한다.   '~는 -와 같다'라는 표현은 틀린 표현이다.

#### 3. 가설검정을 할 때 자주 하는 실수  
* p 값이 작으면 차이가 크다고 말할 수 있다.  
  -> p 값이 작더라도 차이가 크다고 말할 수 없다. 단일표본 t검정에서 t값 계산식을 보면, t값 = (표본평균-비교대상값) / (표준편차 / sqrt(샘플사이즈)) 이다.
  분자에 있는 (표본평균-비교대상값)이 커질수록 차이가 크다고 볼 수 있지만 t값은 이외의 요소도 포함하고 있어, 표준편차가 작거나 샘플사이즈가 크면 t값이 커진다. 따라서 p값만 보고 실험 결과를 해석하는 것은 위험하고 원래 데이터의 평균값을 확인하거나 상자그림이나 바이올린플롯을 활용하는 등의 방법으로 실제와 해석의 차이를 줄일 수 있다.

* p 값이 0.05보다 크므로 차이가 없다고 말할 수 있다.  
  -> 귀무가설이 틀릴 확률은 p값이라는 지표로 컨트롤 가능하지만 귀무가설이 옳을 확률은 컨트롤할 수 없기 때문에 'p값이 0.05보다 크므로 차이가 없다고 말할 수 있다'는 잘못된 해석이다.

* (1-p) 값은 대립가설이 올바를 확률이다.  
  -> 앞의 잘못과 비슷한 맥락으로 틀린 해석이다.




=> 이러한 것들은 통계적가설검정이라는 방식의 한계를 나타낸다.


#### 4. 제1종 오류와 제2종 오류
* 제1종 오류 : 귀무가설이 옳지만 잘못하여 귀무가설을 기각하는 경우
* 제2종 오류 : 귀무가설이 틀렸지만 잘못하여 귀무가설을 채택하는 경우

통계적가설검정에서는 제1종 오류를 범할 확률을 p값으로 계산함. p값이 충분히 작으면 제1종 오류를 범할 확률이 낮아져서 귀무가설을 기각해도 괜찮다고 판단하는 것.

#### 5. 검정의 비대칭성
제1종 오류에 대해서는 그 확률을 컨트롤할 수 있지만, 제2종 오류에 대해서는 컨트롤할 수 없어서 발생하는 것이 검정의 비대칭성

p값 계산 시뮬레이션 할 때 귀무가설이 올바르다고 가정해서 모집단분포를 추정한 표본추출 시뮬레이션으로 수행한 것을 생각해볼 것 -> 귀무가설이 올바르다고 가정했을 때 그것이 틀렸을 확률은 이 틀 안에서 계산 가능하지만 그 반대는 귀무가설이 틀렸다고 가정해야 하기 때문에 같은 방식으로 시뮬레이션하기 어렵다.

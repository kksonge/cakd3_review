### 2집단 데이터에 대한 t검정 - 독립표본 t검정
* 두 모집단으로 부터 표본을 독립적으로 추출해 평균이 같은지 판단
* 1집단 t검정과 비슷하지만 변수가 2개이므로 표준오차와 자유도 계산이 좀 더 복잡함
* Welch 근사법으로 p값을 계산하여 Welch 검정이라고 함

### t검정 구현
학습을 위해 대응표본 t검정과 같은 데이터셋을 사용했지만, 실제로는 데이터에 맞춰 최적의 검정 방법을 골라야 하고 독립표본 t검정은 독립표본 데이터에 적용해야 한다.

```python
# 수치 계산을 위한 라이브러리
import numpy as np
import pandas as pd
from scipy import stats

# 그래프를 그리기 위한 라이브러리
from matplotlib import pyplot as plt
import seaborn as sns
sns.set()

# 표시 자릿수 지정
%precision 3

# 그래프를 주피터 노트북에 그리기 위한 설정
%matplotlib inline
```

```python
paired_test_data = pd.read_csv('/content/drive/MyDrive/python study/Statistics/dataset/3-9-1-paired-t-test.csv')

before = paired_test_data.query('medicine== "before"')['body_temperature']
after = paired_test_data.query('medicine == "after"')['body_temperature']

before = np.array(before)
after = np.array(after)
```

```python
# 평균값
mean_bef = np.mean(before)
mean_aft = np.mean(after)

# 분산
sigma_bef = np.var(before, ddof = 1)
sigma_aft = np.var(after, ddof = 1)

# 샘플사이즈
m = len(before)
n = len(after)

# t값
t_value = (mean_aft - mean_bef) / np.sqrt((sigma_bef/m + sigma_aft/n))
t_value
```

```python
stats.ttest_ind(after, before, equal_var = False)

## p값이 0.05보다 작으므로 유의미한 차이가 있다고 할 수 있다.
## p값이 대응표본 t검정의 결과와 차이가 있다.(0.04 정도) -> 같은 데이터에 대해 같은 목적의 검정을 해도 검정 방법이 다르므로 p값이 달라짐
```

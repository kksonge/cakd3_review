##### 문자 개수 세기(count)



```python
a = 'hobby'
a.count('b')
```

##### 위치 알려주기(find)

```python
a = 'Python is the best choice'
a.find('b')     # b가 처음 나온 위치 반환
a.find('k')     # 존재하지 않으면 -1 반환
```

위치 알려주기(index)

```python
a = 'Life is too short'
a.index('t')    # t가 처음 나온 위치 반환
a.index('k')    # find와 다르게 존재하지 않으면 오류 발생
```

##### 문자열 삽입(join)

```python
','.join('abcd')    # 각각의 문자 사이에 ',' 삽입  
```

```python
','.join(['a', 'b', 'c', 'd'])    # 리스트나 튜플도 입력 가능
```

##### 소문자를 대문자로 바꾸기(upper)

```python
a = 'hi'
a.upper()
```

##### 대문자를 소문자로 바꾸기(lower)

```python
a = 'HI'
a.lower()
```

##### 왼쪽 공백 지우기(lstrip)

```python
a = '  hi '
a.lstrip()
```

##### 오른쪽 공백 지우기(rstrip)

```python
a = '  hi '
a.rstrip()
```

##### 양쪽 공백 지우기(strip)

```python
a = '  hi '
a.strip()
```

##### 문자열 바꾸기(replace)

```python
a = 'Life is too short'
a.replace('Life', 'Your leg')
```

##### 문자열 나누기(split)

```python
a = 'Life is too short'
a.split()    # 공백을 기준으로 문자열 나눔
```

```python
b = 'a:b:c:d'
b.split(':')
```

##### 딕셔너리 관련 함수

###### keys()

```python
# 메모리 낭비를 막기 위해 리스트가 아닌 dic_keys 객체로 반환
a = {'name':'pey', 'phone':'0119993323', 'birth':'1118'}
a.keys()
```

```python
for k in a.keys():
  print(k)
```

```python
list(a.keys())
```

###### values()

```python
a.values()
```

###### items()

```python
a.items()
```

```python
for item in a.items():
  print(item)
```

```python
list(a.items())
```

###### clear()

```python
# 빈 딕셔너리 표현: {}
a.clear()
a
```

###### get()

```python
a = {'name':'pey', 'phone':'0119993323', 'birth':'1118'}
print(a.get('name'))
print(a.get('phone'))
```

```python
# a[key], a.get(key) 기본적으로 같은 결괏값 반환하지만 존재하지 않는 키로 값을 가져오려고 할 때 다름
print(a.get('nokey'))    # None 반환
print(a['nokey'])    # 오류 발생
```

```python
# 찾으려는 Key 값이 없을 경우 디폴트 값을 임의로 정하여 대신 가져오게 함
a.get('foo', 'bar')
```

###### in()

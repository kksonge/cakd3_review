```python
import pandas as pd
import numpy as np
```

```python
a2 = np.random.randint(1, 10, size=(5, 5))
print(a2)
np.save('a', a2)
```

```python
!ls
```

```python
b2 = np.random.randint(1, 10, size=(5, 5))
print(b2)
np.savez('ab', a2, b2)
```

```python
!ls
```

```python
npy = np.load('a.npy')
print(npy)
```

```python
npz = np.load('ab.npz')
print(npz.files)
print(npz['arr_0'])
print(npz['arr_1'])
```

```python
print(a2)
np.savetxt('a.csv', a2, delimiter=',')    # 사람이 읽을 수 있는 text 형태로 저장
```

```python
!ls
```

```python
!cat a.csv    # text로 저장해서 데이터 형태는 반영안되고 실수 형태로 출력됨. binary는 데이터 형태 저장 가능
```

```python
csv = np.loadtxt('a.csv', delimiter=',')
print(csv)
```

```python
print(b2)
np.savetxt('b.csv', b2, delimiter=',', fmt='%.2e', header='c1, c2, c3, c4, c5')
```

```python
!cat b.csv
```

```python
csv = np.loadtxt('b.csv', delimiter=',')
print(csv)
```

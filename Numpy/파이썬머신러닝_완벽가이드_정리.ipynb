### reshape

```python
array1 = np.arange(10)
print('array1:\n', array1, '\n')

array2 = array1.reshape(2,5)
print('array2:\n', array2, '\n')

array3 = array1.reshape(5,2)
print('array3:\n', array3)
```

### Slicing

```python
array1 = np.arange(start=1, stop=10)
array3 = array1[0:3]
print(array1)
print(array3)
print(type(array3))

array4 = array3.tolist()
print(array4)
print(type(array4))
```

```python
array1 = np.arange(start=1, stop=10)
array4 = array1[:3]
print(array4)

array5 = array1[3:]
print(array5)

array6 = array1[:]
print(array6)
```

### fancy indexing

```python
array1d = np.arange(start=1, stop=10)
array2d = array1d.reshape(3,3)

array3 = array2d[[0,1], 2]
print('array2d[[0,1], 2] =>', array3.tolist())

array4 = array2d[[0,1], 0:2]
print('array2d[[0,1], 0:2] =>', array4.tolist())

array5 = array2d[[0,1]]
print('array2d[[0,1]] =>', array5.tolist())
```

### sort(), argsort()

```python
org_array = np.array([3, 1, 5, 9])
print('ì›ë³¸ í–‰ë ¬:', org_array)

# np.sort()ë¡œ ì •ë ¬ ===> ì›ë³¸ì€ ì•ˆë°”ë€œ
sort_array1 = np.sort(org_array)
print('np.sort() í˜¸ì¶œ í›„ ë°˜í™˜ëœ ì •ë ¬ í–‰ë ¬:', sort_array1)
print('np.sort() í˜¸ì¶œ í›„ ì›ë³¸ í–‰ë ¬:', org_array)

# ndarray.sort()ë¡œ ì •ë ¬  ===> ì›ë³¸ì´ ë°”ë€œ
sort_array2 = org_array.sort()
print('org_array.sort() í˜¸ì¶œ í›„ ë°˜í™˜ëœ í–‰ë ¬:', sort_array2)
print('org_array.sort() í˜¸ì¶œ í›„ ì›ë³¸ í–‰ë ¬:', org_array)
```

```python
sort_array1_desc = np.sort(org_array)[::-1]
print('ë‚´ë¦¼ì°¨ìˆœìœ¼ë¡œ ì •ë ¬:', sort_array1_desc)
```

```python
array2d = np.array([[8, 12],
                    [7, 1]])
sort_array2d_axis0 = np.sort(array2d, axis=0)
print('ë¡œìš° ë°©í–¥ìœ¼ë¡œ ì •ë ¬:\n', sort_array2d_axis0)

sort_array2d_axis1 = np.sort(array2d, axis=1)
print('ì»¬ëŸ¼ ë°©í–¥ìœ¼ë¡œ ì •ë ¬:\n', sort_array2d_axis1)
```

```python
import numpy as np

name_array = np.array(['John', 'Mike', 'Sarah', 'Kate', 'Samuel'])
score_array = np.array([78, 95, 84, 98, 88])

sort_indices_asc = np.argsort(score_array)
print('ì„±ì  ì˜¤ë¦„ì°¨ìˆœ ì •ë ¬ ì‹œ score_arrayì˜ ì¸ë±ìŠ¤:', sort_indices_asc)
print('ì„±ì  ì˜¤ë¦„ì°¨ìˆœìœ¼ë¡œ name_arrayì˜ ì´ë¦„ ì¶œë ¥:', name_array[sort_indices_asc])
```
